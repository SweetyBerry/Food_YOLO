```python
import os
import pandas as pd

# taiwan100 資料夾路徑
folder_path = 'taiwan100/train'

# 獲取101個資料夾名稱
food_names = [name for name in os.listdir(folder_path) if os.path.isdir(os.path.join(folder_path, name))]

# 建立 DataFrame，包含三個欄位
data = {
    '食物名稱': food_names,
    '熱量(kcal)': [None] * len(food_names),  # 預設為空
    '修正熱量(kcal)': [None] * len(food_names),  # 預設為空
    '反式脂肪(mg)': [None] * len(food_names),  # 預設為空
    '總碳水化合物(g)': [None] * len(food_names),  # 預設為空
    '每單位重(g)': [None] * len(food_names),  # 預設為空
    '鈉(mg)': [None] * len(food_names)    # 預設為空
}
df = pd.DataFrame(data)

# 儲存為 CSV 檔案
csv_path = 'Calories101.csv'
df.to_csv(csv_path, index=False, encoding='utf-8-sig')

print(f"CSV 檔案已生成: {csv_path}")
```
